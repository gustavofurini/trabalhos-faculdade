**FUNÇÃO SWAP**

\begin{equation}
\text{Swap} = \lambda p. (p \lambda x\,y. (y\,x))
\end{equation}


**EXEMPLO 1:**

\begin{equation}
\text{Swap} = \lambda p. (p \lambda 10\,20. (20\,10))
\end{equation}
\begin{equation}
\text{Swap}(10, 20) \rightarrow (20, 10)
\end{equation}

**EXEMPLO 2:**

\begin{equation}
\text{Swap} = \lambda p. (p \lambda 30\,40. (40\,30))
\end{equation}
\begin{equation}
\text{Swap}(30, 40) \rightarrow (40, 30)
\end{equation}

**EXEMPLO 3:**

\begin{equation}
\text{Swap} = \lambda p. (p \lambda 50\,60. (60\,50))
\end{equation}
\begin{equation}
\text{Swap}(50, 60) \rightarrow (60, 50)
\end{equation}

```python
def swap(p):
    x, y = p
    return (y, x)

# EXEMPLO 1
par1 = (10, 20)
parTrocado = swap(par1)

print(f"PAR ORDENADO ORIGINAL {par1}")
print(f"PAR ORDENADO TROCADO {parTrocado} \n")

# EXEMPLO 2
par2 = (30, 40)
parTrocado = swap(par2)

print(f"PAR ORDENADO ORIGINAL {par2}")
print(f"PAR ORDENADO TROCADO {parTrocado} \n")

# EXEMPLO 3
par3 = (50, 60)
parTrocado = swap(par3)

print(f"PAR ORDENADO ORIGINAL {par3}")
print(f"PAR ORDENADO TROCADO {parTrocado} \n")
```

**FUNÇÃO AO PAR**
\begin{equation}
AoPar = \lambda pf. \left( p \left( \lambda xy. (fx)(fy) \right) \right)
\end{equation}

**EXEMPLO 1:**

\begin{align}
AoPar((10,20),(λx.x^2)) &= ((10,20)(λxy.((λx.x^2)x)((λx.x^2)y))) \\
&= (λxy.((λx.x^2)x)((λx.x^2)y))\,10\ 20
\end{align}
\
\begin{align}
&= ((λxy.((λx.x^2)x)((λx.x^2)y))\,10\ 20) \\
&= (((λxy.(x^2)(y^2))\,10\ 20))
\end{align}
\
\begin{align}
&= ((λxy.(x^2)(y^2))\,10  20) \\
&= ((10^2)(20^2)) \\
&= (100\ 400) \\
&= (100, 400)
\end{align}

**EXEMPLO 2:**

\begin{align}
AoPar((30,40),(λx.x^2)) &= ((30,40)(λxy.((λx.x^2)x)((λx.x^2)y))) \\
&= (λxy.((λx.x^2)x)((λx.x^2)y))\,30\ 40
\end{align}
\
\begin{align}
&= ((λxy.((λx.x^2)x)((λx.x^2)y))\,30\ 40) \\
&= (((λxy.(x^2)(y^2))\,30\ 40))
\end{align}
\
\begin{align}
&= ((λxy.(x^2)(y^2))\,30\  40) \\
&= ((30^2)(40^2)) \\
&= (900\ 1600) \\
&= (900, 1600)
\end{align}

**EXEMPLO 3:**

\begin{align}
AoPar((50,60),(λx.x^2)) &= ((50,60)(λxy.((λx.x^2)x)((λx.x^2)y))) \\
&= (λxy.((λx.x^2)x)((λx.x^2)y))\,50\ 60
\end{align}
\
\begin{align}
&= ((λxy.((λx.x^2)x)((λx.x^2)y))\,50\ 60) \\
&= (((λxy.(x^2)(y^2))\,50\ 60))
\end{align}
\
\begin{align}
&= ((λxy.(x^2)(y^2))\,50  60) \\
&= ((50^2)(60^2)) \\
&= (2500\ 3600) \\
&= (2500, 3600)
\end{align}







```python
def ao_par(f, p):
    return (f(p[0]), f(p[1]))

# EXEMPLO 1
par1 = (10, 20)
aoquadrado = lambda x: x**2
resultado = ao_par(aoquadrado, par1)

print(f"PAR ORDENADO ORIGINAL {par1}")
print(f"PAR ORDENADO AO QUADRADO {resultado} \n")

# EXEMPLO 2
par2 = (30, 40)
aoquadrado = lambda x: x**2
resultado = ao_par(aoquadrado, par2)

print(f"PAR ORDENADO ORIGINAL {par2}")
print(f"PAR ORDENADO AO QUADRADO {resultado} \n")

# EXEMPLO 3
par3 = (50, 60)
aoquadrado = lambda x: x**2
resultado = ao_par(aoquadrado, par3)

print(f"PAR ORDENADO ORIGINAL {par3}")
print(f"PAR ORDENADO AO QUADRADO {resultado} \n")
```

**FUNÇÃO SOMAPAR**

\begin{equation}
SomaPar = \lambda p_1 p_2. (p_1 (\lambda xy. (p_2 (\lambda uv. (x + u)(y + v)))))
\end{equation}

**EXEMPLO 1:**

\begin{align}
SomaPar((10,20),(30,40)) &= ((10,20)(\lambda xy.((30,40)(\lambda uv.(x+u)(y+v))))) \\
&= (\lambda xy.((30,40)(\lambda uv.(x+u)(y+v))))(10,20)
\end{align}
\
\begin{equation}
= (((30,40)(\lambda uv.(10+u)(20+v))))
\end{equation}
\
\begin{align}
&= ((\lambda uv.(10+u)(20+v))30 \ 40) \\
&= ((10+30)(20+40)) \\
&= (40,60)
\end{align}

**EXEMPLO 2:**

\begin{align}
SomaPar((50,60),(70,80)) &= ((50,60)(\lambda xy.((70,80)(\lambda uv.(x+u)(y+v))))) \\
&= (\lambda xy.((70,80)(\lambda uv.(x+u)(y+v))))(50,60)
\end{align}
\
\begin{equation}
= (((70,80)(\lambda uv.(50+u)(60+v))))
\end{equation}
\
\begin{align}
&= ((\lambda uv.(50+u)(60+v))70 \ 80) \\
&= ((50+70)(60+80)) \\
&= (120,140)
\end{align}

**EXEMPLO 3:**

\begin{align}
SomaPar((90,100),(110,120)) &= ((90,100)(\lambda xy.((110,120)(\lambda uv.(x+u)(y+v))))) \\
&= (\lambda xy.((110,120)(\lambda uv.(x+u)(y+v))))(90,100)
\end{align}
\
\begin{equation}
= (((110,120)(\lambda uv.(90+u)(110+v))))
\end{equation}
\
\begin{align}
&= ((\lambda uv.(90+u)(100+v))110 \ 120) \\
&= ((90+110)(100+120)) \\
&= (200,220)
\end{align}



```python
def soma_par(p1, p2):
    return (p1[0] + p2[0], p1[1] + p2[1])

# EXEMPLO 1
par1 = (10, 20)
par2 = (30, 40)
resultado = soma_par(par1, par2)

print("PAR 1:", par1)
print("PAR 2:", par2)
print(f"SOMA DOS PARES: {resultado} \n")

# EXEMPLO 2
par3 = (50, 60)
par4 = (70, 80)
resultado = soma_par(par3, par4)

print("PAR 3:", par3)
print("PAR 4:", par4)
print(f"SOMA DOS PARES: {resultado} \n")

# EXEMPLO 3
par5 = (90, 100)
par6 = (110, 120)
resultado = soma_par(par5, par6)

print("PAR 5:", par5)
print("PAR 6:", par6)
print(f"SOMA DOS PARES: {resultado} \n")
```
