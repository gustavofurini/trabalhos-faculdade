```python

def dfs_iterative(graph, source_node, destination_node):
  visited = [] # Lista de vertices ja visitados
  stack = [] # Pilha para controlar a ordem de visitação dos vértices

  stack.append(source_node) # adiciona nó de origem na pilha
  while len(stack) > 0: # executa enquanto a pilha não estiver vazia
    element = stack.pop() # tira o elemento que está no topo da pilha
    if element not in visited: # caso o elemnto não esteja em 'visited'
      visited.append(element) # adiciona elemento na lista de visitado
      # adiciona na pilha, os vizinhos do elemnto que ainda não foram visitado
      # para seguir a ordem alfabetica, ordenamos os vizinhos da maneira reversa
      for adj in sorted(graph[element], reverse=True):
        if adj not in visited:
          stack.append(adj)
        if adj == destination_node:
          return visited
  return visited


```

```python
def dfs_recursive(graph, current_node, destination_node, visited=None):
    if visited is None:
        visited = set()  # Usando um conjunto para armazenar nós visitados

    visited.add(current_node)  # Marca o nó atual como visitado

    # Verifica se o nó atual é o destino
    if current_node == destination_node:
        return [current_node]

    # Explora os vizinhos do nó atual recursivamente
    for neighbor in sorted(graph[current_node], reverse=True):
        if neighbor not in visited:
            path = dfs_recursive(graph, neighbor, destination_node, visited)
            if path:  # Se um caminho para o destino for encontrado
                return [current_node] + path

    # Se não encontrar o destino a partir deste nó, retorna vazio
    return []


# Chamada da função para encontrar o caminho de "A" para "E"
path = dfs_recursive(G1, "A", "E")
print("Caminho encontrado:", path)
```

```python

def bfs_iterative(graph, source_node, destination_node):
  visited = [] # Lista de vertices ja visitados
  queue = [] # Fila para controlar a ordem de visitação dos vértices

  queue.append(source_node) # adiciona nó de origem na fila
  while len(queue) > 0: # executa enquanto a fila não estiver vazia
    element = queue.pop(0) # tira o elemento que está no inicio da fila
    if element not in visited: # caso o elemnto não esteja em 'visited'
      visited.append(element) # adiciona elemento na lista de visitado
      # adiciona na fila, os vizinhos do elemnto que ainda não foram visitado
      for adj in sorted(graph[element]):
        if adj not in visited:
          queue.append(adj)
        if adj == destination_node:
          return visited
  return visited




bfs_iterative(G1, "A", "E")
```
